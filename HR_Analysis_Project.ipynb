## **HR Analytics: Employee Attrition Analysis and Interactive Dashboard Using Python & Power BI**

## **Business Problem**

**Employee attrition is a major challenge for organizations as it affects productivity, increases recruitment costs, and impacts overall business performance. The HR department requires data-driven insights to identify the factors responsible for employee turnover and understand which groups of employees are more likely to leave the organization. By analyzing HR data, suitable strategies can be developed to improve employee retention and workplace satisfaction.**

## **Objective**

**The main objective of this project is to analyze employee data using Python and identify the key factors influencing employee attrition. The project aims to perform data cleaning, exploratory data analysis, and visualization to discover meaningful patterns related to employee satisfaction, salary, department, working hours, promotions, and experience. The insights obtained from the analysis will support HR professionals in making informed decisions to reduce employee turnover.**
